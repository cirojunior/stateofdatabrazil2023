**Tratando o nome das colunas do DataFrame para ficar melhor o uso.**


Crio uma função para reaproveitar código e dentro dele coloco um for para passar por todas as colunas que existem no DataFrame e na sequencia uso um Try para garantir que erros não vão deixar o código parar, dentro do Try faço os replace "Substituição de caracteres" e o split "Para separar os textos por um separador e transformar em lista".


Utilizo do rename para renomear as colunas do DataFrame onde passo a variável do for como o nome original da coluna e a variável que contem o texto formatado.


Uso excep pass para seguir o código caos encontre um problema, uma vez alterado o nome da coluna se rodar o codigo novamente vai dar erro porque não vai encontrar o nome original que veio dos dados.

```python
def replaceColumns(coluna):
  for i in coluna:
    try:
      c = i.replace("('","")
      c = c.replace("')","")
      c = c.replace("', '",",")
      c = c.replace(" ,",",")
      c = c.split(",")
      df.rename(columns={i:c[1]}, inplace=True)
    except:
      pass
    #df.rename(columns={i:c[1]}, inplace=True)

replaceColumns(df.columns)
```
