**Tratando o nome das colunas do DataFrame para ficar melhor o uso.**


Crio uma função para reaproveitar código e dentro dele coloco um for para passar por todas as colunas que existem no DataFrame e na sequencia uso um Try para garantir que erros não vão deixar o código parar, dentro do Try faço os replace "Substituição de caracteres" e o split "Para separar os textos por um separador e transformar em lista".


Utilizo do rename para renomear as colunas do DataFrame onde passo a variável do for como o nome original da coluna e a variável que contem o texto formatado.


Uso excep pass para seguir o código caos encontre um problema, uma vez alterado o nome da coluna se rodar o codigo novamente vai dar erro porque não vai encontrar o nome original que veio dos dados.

```python
def replaceColumns(coluna):
  for i in coluna:
    try:
      c = i.replace("('","")
      c = c.replace("')","")
      c = c.replace("', '",",")
      c = c.replace(" ,",",")
      c = c.split(",")
      df.rename(columns={i:c[1]}, inplace=True)
    except:
      pass
    #df.rename(columns={i:c[1]}, inplace=True)

replaceColumns(df.columns)
```

```python
# prompt: quero criar um filtro no grafico de cima para poder filtrar  a Faixa idade da pesquisa

import matplotlib.pyplot as plt
def filter_idade_genero(df, faixa_idade=None):
  """
  Filtra o DataFrame 'idadeGenero' por faixa de idade e gera o gráfico.

  Args:
    df: DataFrame original contendo os dados.
    faixa_idade: String com a faixa de idade para filtrar (e.g., '25-29').
                 Se None, mostra todos os dados.
  """
  idadeGenero = df.groupby(['Faixa idade','Genero']).size().reset_index(name='Quantidade')

  if faixa_idade:
    idadeGenero_filtered = idadeGenero[idadeGenero['Faixa idade'] == faixa_idade]
    if idadeGenero_filtered.empty:
      print(f"Nenhum dado encontrado para a faixa etária: {faixa_idade}")
      return
    plot_df = idadeGenero_filtered.groupby('Genero').size()
  else:
    plot_df = idadeGenero.groupby('Genero').size()

  plot_df.plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
  plt.gca().spines[['top', 'right',]].set_visible(False)
  plt.title(f"Distribuição de Gênero por Faixa Etária: {faixa_idade if faixa_idade else 'Todas'}")
  plt.xlabel('Quantidade')
  plt.ylabel('Gênero')
  plt.show()

# Exemplo de uso:
# Para filtrar por uma faixa de idade específica (substitua '25-29' pela faixa desejada)
# filter_idade_genero(df, faixa_idade='25-29')

# Para mostrar todos os dados (sem filtro)
filter_idade_genero(df, '22-24')
```
